# Part 5: Policy Improvement Theorem

## Q5: Formal Proof

**Prove that the ε-greedy policy π' derived from q_π, where π is any ε-soft policy, satisfies:**

$$v_{\pi'}(s) \geq v_\pi(s) \text{ for all } s \in S$$

---
# DEFINITIONS

**ε-soft policy π:**

A policy where every action has minimum probability in every state:
$$\pi(a|s) \geq \frac{\varepsilon}{|A|} \text{ for all } s, a$$

**ε-greedy policy π' with respect to q_π:**

For the greedy action $a^* = \arg\max_a q_\pi(s,a)$:
$$\pi'(a^*|s) = 1 - \varepsilon + \frac{\varepsilon}{|A|}$$
$$\pi'(a|s) = \frac{\varepsilon}{|A|} \text{ for } a \neq a^*$$

where $|A|$ is the number of actions.

---
# PROOF

## Step 1: Express $v_\pi$ in terms of $q_\pi$

By definition of state value under policy π:
$$v_\pi(s) = \sum_a \pi(a|s) q_\pi(s,a)$$

This is the expected Q-value weighted by the policy distribution.

## Step 2: Express $v_{\pi'}$ in terms of $q_\pi$

Similarly, for policy π':
$$v_{\pi'}(s) = \sum_a \pi'(a|s) q_\pi(s,a)$$

The ε-greedy policy places distribution π' over actions.

## Step 3: Key Insight - π' concentrates probability on better actions

By construction of ε-greedy:
- Greedy action $a^*$: $\pi'(a^*|s) = 1 - \varepsilon + \frac{\varepsilon}{|A|}$ (highest probability)
- Other actions: $\pi'(a|s) = \frac{\varepsilon}{|A|}$ (lower probability)

where $a^* = \arg\max_a q_\pi(s,a)$

## Step 4: Compare $v_{\pi'}(s)$ and $v_\pi(s)$

$$v_{\pi'}(s) = \sum_a \pi'(a|s) q_\pi(s,a)$$

Separate the greedy action:
$$v_{\pi'}(s) = \pi'(a^*|s) \cdot q_\pi(s,a^*) + \sum_{a \neq a^*} \pi'(a|s) q_\pi(s,a)$$

Substitute the ε-greedy probabilities:
$$v_{\pi'}(s) = \left(1 - \varepsilon + \frac{\varepsilon}{|A|}\right) q_\pi(s,a^*) + \sum_{a \neq a^*} \frac{\varepsilon}{|A|} q_\pi(s,a)$$

$$= q_\pi(s,a^*) - \varepsilon q_\pi(s,a^*) + \frac{\varepsilon}{|A|} q_\pi(s,a^*) + \frac{\varepsilon}{|A|} \sum_{a \neq a^*} q_\pi(s,a)$$

$$= q_\pi(s,a^*) + \frac{\varepsilon}{|A|} \left[q_\pi(s,a^*) + \sum_{a \neq a^*} q_\pi(s,a)\right] - \varepsilon q_\pi(s,a^*)$$

$$= q_\pi(s,a^*) + \frac{\varepsilon}{|A|} \sum_a q_\pi(s,a) - \varepsilon q_\pi(s,a^*)$$

## Step 5: Key Lemma - Greedy action is at least as good as average

Since $a^* = \arg\max_a q_\pi(s,a)$:
$$q_\pi(s,a^*) \geq q_\pi(s,a) \text{ for all } a$$

Therefore:
$$q_\pi(s,a^*) \geq \frac{1}{|A|} \sum_a q_\pi(s,a)$$

Multiply by $|A|$:
$$|A| \cdot q_\pi(s,a^*) \geq \sum_a q_\pi(s,a)$$

## Step 6: Apply the lemma to Step 4 result

From Step 4:
$$v_{\pi'}(s) = q_\pi(s,a^*) + \frac{\varepsilon}{|A|} \sum_a q_\pi(s,a) - \varepsilon q_\pi(s,a^*)$$

From Step 5 lemma:
$$\sum_a q_\pi(s,a) \leq |A| \cdot q_\pi(s,a^*)$$

Therefore:
$$v_{\pi'}(s) \geq q_\pi(s,a^*) + \frac{\varepsilon}{|A|} \cdot |A| \cdot q_\pi(s,a^*) - \varepsilon q_\pi(s,a^*)$$

$$= q_\pi(s,a^*) + \varepsilon q_\pi(s,a^*) - \varepsilon q_\pi(s,a^*)$$

$$= q_\pi(s,a^*)$$

## Step 7: Relate $q_\pi(s,a^*)$ to $v_\pi(s)$

We have:
$$v_\pi(s) = \sum_a \pi(a|s) q_\pi(s,a)$$

Consider the improvement:
$$q_\pi(s,a^*) - v_\pi(s) = q_\pi(s,a^*) - \sum_a \pi(a|s) q_\pi(s,a)$$

$$= \sum_a \pi(a|s) [q_\pi(s,a^*) - q_\pi(s,a)]$$

Since $a^* = \arg\max_a q_\pi(s,a)$:
$$q_\pi(s,a^*) \geq q_\pi(s,a) \text{ for all } a$$

And since π is a valid probability distribution: $\pi(a|s) \geq 0$ for all $a$ 

Therefore:
$$\sum_a \pi(a|s) [q_\pi(s,a^*) - q_\pi(s,a)] \geq 0$$

This gives us:
$$q_\pi(s,a^*) \geq v_\pi(s)$$

## Step 8: Combine Results

From Step 6:  $v_{\pi'}(s) \geq q_\pi(s,a^*)$

From Step 7:  $q_\pi(s,a^*) \geq v_\pi(s)$

By transitivity:
$$\boxed{v_{\pi'}(s) \geq v_\pi(s)}$$

---
# CONCLUSION

**Theorem:** For any state $s \in S$:
$$v_{\pi'}(s) \geq v_\pi(s)$$

where π is an ε-soft policy and π' is the ε-greedy policy derived from $q_\pi$.

**This proves that the ε-greedy policy is at least as good as the original ε-soft policy at every state.**

---
# INTUITION

**Why does this theorem hold?**

1. **Exploitation beats exploration:** The ε-greedy policy concentrates probability on the action with the highest estimated Q-value (the greedy action).

2. **Better than average:** By choosing the best-valued action, you're guaranteed at least the value of that action, which is at least as good as the average weighted by the original policy.

3. **Still explores:** The ε-soft property ensures you still explore all actions (due to the $ε/|A|$ probability on non-greedy actions).

4. **Foundation of RL:** This principle underlies:
   - Policy Iteration algorithms
   - Q-learning
   - Expected SARSA
   - Actor-Critic methods

---
# RELATED CONCEPTS

**Policy Improvement Theorem (Deterministic):**

If $q_\pi(s, \pi'(s)) \geq v_\pi(s)$ for all $s$, then $v_{\pi'}(s) \geq v_\pi(s)$ for all $s$.

The current proof is the **stochastic extension** of this theorem for ε-soft policies.

**Generalized Policy Iteration (GPI):**

The theorem enables the GPI algorithm:
1. **Policy Evaluation:** Compute $q_\pi$ (or approximate it)
2. **Policy Improvement:** Update π to be ε-greedy w.r.t. $q_\pi$
3. **Repeat:** Convergence guaranteed

This cycle is the basis of many RL algorithms.
